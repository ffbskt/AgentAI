```python
import os
from pathlib import Path

# Change these paths if your repo is stored elsewhere in Drive.
REPO_ROOT = Path('/content/drive/MyDrive/Transformer_Graph3')
BUNDLE_DIR = REPO_ROOT / 'colab_bundle'
RUN_ROOT = REPO_ROOT / 'colab_runs' / 'lstm_32k_curriculum'
CHECKPOINT_DIR = RUN_ROOT / 'checkpoints'
RUNS_DIR = RUN_ROOT / 'runs'

CHECKPOINT_DIR.mkdir(parents=True, exist_ok=True)
RUNS_DIR.mkdir(parents=True, exist_ok=True)

print('Bundle dir:', BUNDLE_DIR)
print('Run root:', RUN_ROOT)
```

```python
import os

os.chdir(BUNDLE_DIR)
print('Working dir:', os.getcwd())
```

```python
MODEL_KIND = 'lstm'
DROPOUT = '0.01'
DEVICE = 'cpu'  # Change to 'cuda' in Colab GPU runtime if available.
EXTRA_ARGS = ''

cmd = f'''python train.py \
  --kind {MODEL_KIND} \
  --device {DEVICE} \
  --dropout {DROPOUT} \
  --save-dir "{CHECKPOINT_DIR}" \
  --save-each-digit \
  --protocol-file "{RUNS_DIR / 'protocol.jsonl'}" \
  {EXTRA_ARGS} \
  | tee "{RUNS_DIR / 'train.log'}"'''

print(cmd)
```
